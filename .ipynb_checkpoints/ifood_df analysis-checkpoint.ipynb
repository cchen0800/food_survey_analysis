```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ifood_df.csv")
```

```python
df.head(15)
```

```python
## Checking for missing values

missing_values = df.isnull().sum()

missing_values[missing_values >0]
```

```python
# No missing values!
# lets summarize the table
df.describe()
```

```python
## Observations

# Average income is around $51,622 but ranges from $1730 to $113,734

# Most customers have either one or no child and teenager at home

# Customers spend most on wines followed by met products. 

# Most customers have not accepted any offers, there is a large STD with some customers accepting up 
# to 4 Campaigns

# Column MntRegularProds has a negative value which is impossible since spending can't be negative

```

```python
# Investigate MntRegularProds

# Filtering rows where MntRegularProds is negative

negative_values = df[df['MntRegularProds'] < 0]
negative_values
```

```python
# checking to see if MntTotal - MntGold gives us MntRegularProds amount

negative_values['Expected_Amount'] = negative_values['MntTotal'] - negative_values['MntGoldProds']

negative_values[['Expected_Amount','MntRegularProds']]
```

```python
# appears to be calculation error within these rows
# lets make MntRegular Prods the difference between 'MntTotal' and 'MntGoldProds'
```

```python
df.MntRegularProds = df.MntTotal - df.MntGoldProds

df.describe()
```

```python
## Answer appears to still be negative
# May be easier to remove the rows from table considering table size and only 3 errors

df = df[df['MntRegularProds'] >= 0]
```

```python
df.shape
```

```python
#lets visualize distribution of income

sns.set_style('whitegrid')

# Plot the distribution of the Income

plt.figure(figsize=(10,6))
sns.histplot(df.Income, bins = 30, kde = True, color = 'skyblue')
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Distribution of Income is right skewed, suggesting that majority of customers have lower incomes

#Lets visualize distribution of spending on the variety of products

product_columns = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts', "MntSweetProducts", "MntGoldProds"]

```

```python
# Plot the distribution of spending across different product categories


plt.figure(figsize=(15, 15))

for i, product in enumerate(product_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[product], bins=30, kde=True, color="skyblue")
    plt.title(f"Distribution of {product}")
    plt.xlabel("Spending")
    plt.ylabel("Number of Customers")
    plt.tight_layout()

plt.show()
```

```python
# Lets visualize the relationship between Income and MntTotal

plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["Income"], y=df["MntTotal"], alpha=0.6, edgecolor=None)
plt.title("Relationship between Income and Total Spending")
plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.show()
```

```python
# Visualizing relationship between number of kids (Kidhome) and teenagers (Teenhome) on total spending
```

```python
# Plotting the impact of Kidhome and Teenhome on total spending

plt.figure(figsize=(14, 6))

# For Kidhome
plt.subplot(1, 2, 1)
sns.boxplot(x=df.Kidhome, y=df.MntTotal, palette="pastel")
plt.title("Impact of Number of Kids at Home on Total Spending")
plt.xlabel("Number of Kids at Home")
plt.ylabel("Total Spending")

# For Teenhome
plt.subplot(1, 2, 2)
sns.boxplot(df.Teenhome, y=df.MntTotal, palette="pastel")
plt.title("Impact of Number of Teenagers at Home on Total Spending")
plt.xlabel("Number of Teenagers at Home")
plt.ylabel("Total Spending")

plt.tight_layout()
plt.show()
```

```python
# Appears that customers with children at home spend substantially less compared to customers with no children
# Customers with teenagers spend less than no teenagers, but not by significant amount relative to customers with children

# Overall, customers with no children tend to spend more

```

```python
# Lets analyze effect of marital status on total spending
# The Dataset gives us multiple columns that are binary for Divorced, Married, Single, Togethor, and Widow
# Lets convert these wide columns to long

marital_columns = [col for col in df if 'marital_' in col]

#melting df

marital_df = pd.melt(df, id_vars = 'MntTotal', value_vars = marital_columns, var_name = 'Marital Status', value_name = 'Flag')


marital_df['Marital Status'] = marital_df['Marital Status'].replace('marital_','')

marital_df
```

```python
marital_df = marital_df[marital_df["Flag"] == 1]

# Visualizing new df

plt.figure(figsize=(14, 7))
sns.boxplot(x=marital_df["Marital Status"], y=marital_df["MntTotal"], palette="pastel")
plt.title("Impact of Marital Status on Total Spending")
plt.xlabel("Marital Status")
plt.ylabel("Total Spending")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Customers who are single and togethor spend slightly less than divorced and married
# Widows spend the most 
# My hypothesis is that marital status might be correlated with age, which in turn is also correlated with spending habits

#Lets investigate age further
```

```python
plt.figure(figsize=(20, 9))
sns.histplot(df["Age"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# graph is  slightly right skewed with biggest concentration between the ages of 40 and 50

# plotting correlation between age and total spending

plt.figure(figsize=(14, 7))
sns.boxplot(x=pd.cut(df["Age"], bins=[20, 30, 40, 50, 60, 70, 80], 
                    labels=['20s', '30s', '40s', '50s', '60s', '70+']), 
            y=df["MntTotal"], palette="pastel")
plt.title("Distribution of Total Spending by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Spending")
plt.show()
```

```python
# 20's and 70's spend the most with the highest medians
# 30's spends the least (Potentially because of children)
# 70's spends the most
```

```python
# Lets calculate correlation matrix for numeric variables
correlation_matrix = df[["Income", "Recency", "MntWines", "MntFruits", "MntMeatProducts", 
                              "MntFishProducts", "MntSweetProducts", "MntGoldProds", "MntTotal", 
                              "MntRegularProds",'Age','NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', "AcceptedCmpOverall"]].corr()

# Plot the correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Income has highest correlations with spending categories, particuarly wine and meat indicating these are top spend categories
# 
```
